# Demo `plotting.oneD`

This notebook demonstrates how to use the `heputl.plotting.oneD` module to visualize 1D feature distributions for multiple samples.

## 1. Import Required Libraries

We start by importing numpy for data generation and the `plot_feature_hist_for_n_samples` function from the plotting module.

```python
import sys
import os
sys.path.insert(0, os.path.abspath(os.path.join(os.pardir, "src")))
import numpy as np
from heputl.plotting.oneD import plot_feature_hist_for_n_samples
```

initialize also a logger

```python
from heputl import logging as logg

# Set up logging (INFO level by default)
logger = logg.get_logger(__name__)

logger.info('Running demo of oneD plotting.')
```

## 2. Prepare Sample Data

Let's generate synthetic data for two samples. We'll also create sample weights and assign names to each sample.

```python
# Generate synthetic data for two samples
np.random.seed(42)
size = 10000
sample1 = np.random.normal(loc=0, scale=1, size=size)
sample2 = np.random.normal(loc=2, scale=1.5, size=size)

# Optional: create weights for each sample
weights1 = np.random.uniform(0.8, 1.2, size=sample1.shape)
weights2 = np.random.uniform(0.5, 1.5, size=sample2.shape)

# Sample names
sample_names = ['Sample A', 'Sample B']

data = [sample1, sample2]
weights = [weights1, weights2]

logger.info('Data and weights prepared for plotting.')
```

## 3. Plot 1D Feature Histogram for Multiple Samples

Now, let's visualize the feature distributions for both samples using the `plot_feature_hist_for_n_samples` function. We'll use the generated data, sample names, and weights.

```python
plot_feature_hist_for_n_samples(
    data=data,
    sample_names=sample_names,
    weights=weights,
    bins=50,
    xlabel='Feature Value',
    ylabel='Fraction of Events',
    show_plt=True
)
```

## 4. Customize Histogram Appearance

You can customize the histogram by changing the number of bins, enabling/disabling log scale, setting axis labels, and using different color palettes. Below are some examples.

```python
plot_feature_hist_for_n_samples(
    data=data,
    sample_names=sample_names,
    weights=weights,
    bins=80,
    xlabel='Custom X Label',
    ylabel='Event Density',
    ylogscale=False,
    legend_outside=False,
    show_plt=True
)
```

```python
import numpy as np
import pandas as pd
from read_data_output_graph import read_data_output_graph
```

```python
directory = "./protein_data"
pro_elements = ["C"] 
pro_ele_rad = [1.70]

lig_elements = ["C", "N", "O", "S"]
lig_ele_rad = [1.70, 1.55, 1.52, 1.80]
pdbid = "1a99"
cutoff = 10
no_cutoff = False

fname = read_data_output_graph("1a99",pro_elements, pro_ele_rad, directory, lig_elements, lig_ele_rad,cutoff=cutoff,no_cutoff=no_cutoff)
```

```python
from petls import dFlag
```

```python
max_r = cutoff
stepsize = 0.1
num_steps = max_r/stepsize
unflipped = dFlag(filename=fname, max_dim=2)

for i in range(round(num_steps)+1):
    for dim in range(3):
        a = i*stepsize
        b = (i+1)*stepsize
        fname = f"protein_L/dim{dim}_a{a:.2f}_b{b:.2f}.mkt"
        L = unflipped.store_L(dim, a, b, prefix=fname)
        print(f"saved matrix to file {fname}")
# unflipped.time_to_csv("unflipped_profile.csv")
print("finished")
```
